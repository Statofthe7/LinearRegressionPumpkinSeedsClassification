```python
#Importing pandas and scikitlearn libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Importing Data

```python
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("muratkokludataset/pumpkin-seeds-dataset")

# Full path
csv_file_path = os.path.join(path, "Pumpkin_Seeds_Dataset/Pumpkin_Seeds_Dataset.xlsx")

#print("Full path to CSV file:", csv_file_path)

#Reading pumpkin seeds dataset into a dataframe
ds = pd.read_excel(csv_file_path)
print(ds.head(5))
```

```python
#Convert seed type from string to integer
ds['Seed_Type'] = ds['Class'].map({'Çerçevelik':0, 'Ürgüp Sivrisi':1})
print(ds['Seed_Type'].value_counts())
```

```python
#Setting up X and y values
X = ds[['Area', 'Perimeter', 'Solidity', 'Roundness']]
y = ds[['Seed_Type']]
```

Preprocessing: Regression

```python
#Split train and test data in the ratio of 2:1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)

#Transform train and test data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
y_train = sc.fit_transform(y_train)
y_test = sc.transform(y_test)
```

Training: Regression

```python
#Initiate linear regression model
linear_model = LinearRegression()
```

```python
#Feed training data to the model
linear_model.fit(X_train, y_train)
```

Evaluation: Regression

```python
#Predict seed type for test data
y_pred = linear_model.predict(X_test)

#Check regression score function. Best possible score is 1.0
r2 = metrics.r2_score(y_test, y_pred)
```

```python
#Print regression score
print(f'r2 Score: {r2}')
```
